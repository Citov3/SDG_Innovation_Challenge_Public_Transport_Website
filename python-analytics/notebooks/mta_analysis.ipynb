# Python Notebook for cleaning and analyzing MTA station data





```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load raw CSV (replace with your actual file)

```python
mta_df = pd.read_csv('../data/raw/mta_data.csv')
```

# Quick look at data

```python
print(mta_df.head())
print(mta_df.info())
```

# Convert time column to datetime

```python
mta_df['time'] = pd.to_datetime(mta_df['time'], format='%H:%M:%S')
```

# Example: Filter station '34 ST-PENN STA'

```python
mta_penn = mta_df[mta_df['Station'] == '34 ST-PENN STA']
```

# Group by time and sum "busy" column

```python
mta_penn_grouped = mta_penn.groupby('time').agg({'busy': 'sum'}).sort_values(by='busy', ascending=False)
```

# Resample by hour

```python
mta_hourly = mta_penn_grouped.resample('60T').sum().sort_values(by='busy', ascending=False)
print(mta_hourly.head())
```

# Plot busiest hours

```python
mta_hourly.plot(kind='bar', figsize=(12,5), legend=False, title='Busy Levels at 34 ST-PENN STA')
plt.xlabel('Time')
plt.ylabel('Number of People')
plt.show()
```

# Save cleaned data for backend use

```python
mta_hourly.to_json('../data/processed/mta_penn_hourly.json')
```
