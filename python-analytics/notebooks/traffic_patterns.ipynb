## Python Notebook for analyzing live traffic patterns

```python
import pandas as pd
import json
```

```python
# Load traffic JSON or CSV
import os

# Create dummy data directory if it doesn't exist
output_dir = '/content/data/raw' # Changed to absolute path
os.makedirs(output_dir, exist_ok=True)

# Create dummy traffic data
dummy_data = {
    'time': ['07:00:00', '07:30:00', '08:00:00', '08:30:00', '09:00:00', '17:00:00', '17:30:00', '18:00:00', '18:30:00', '19:00:00'],
    'congestion_level': [5, 7, 8, 9, 6, 8, 9, 7, 6, 5]
}
dummy_df = pd.DataFrame(dummy_data)
dummy_df.to_csv(os.path.join(output_dir, 'traffic_data.csv'), index=False)

traffic_df = pd.read_csv('/content/data/raw/traffic_data.csv') # Changed to absolute path

# Quick overview
print(traffic_df.head())
print(traffic_df.info())
```

# Example: Group by hour to find congestion patterns


```python
# Example: Group by hour to find congestion patterns
traffic_df['time'] = pd.to_datetime(traffic_df['time'], format='%H:%M:%S')
traffic_hourly = traffic_df.groupby(traffic_df['time'].dt.hour).agg({'congestion_level':'mean'})

print(traffic_hourly)
```

```python
# Save processed traffic patterns
traffic_hourly.to_json('../data/processed/traffic_hourly.json')
```
