The following code calculates carbon emissions for various transportation modes and saves the results to a JSON file.

```python
# Carbon emission factors (kg CO2 per km)
EMISSION_FACTORS = {
    "subway": 0.04,
    "bus": 0.12,
    "car": 0.25,
    "bike": 0.0,
    "walk": 0.0
}

def calculate_emissions(mode, distance_km):
    return round(EMISSION_FACTORS.get(mode.lower(), 0.2) * distance_km, 3)

# Example usage
example_routes = [
    {"mode": "subway", "distance": 5},
    {"mode": "bus", "distance": 5},
    {"mode": "car", "distance": 5},
]

for route in example_routes:
    print(f"{route['mode']} ({route['distance']} km) → {calculate_emissions(route['mode'], route['distance'])} kg CO2")

# Save sample data for backend
import pandas as pd
import os

carbon_df = pd.DataFrame(example_routes)
carbon_df['co2_kg'] = carbon_df.apply(lambda row: calculate_emissions(row['mode'], row['distance']), axis=1)

# Create the directory if it doesn't exist
output_dir = '../data/processed'
os.makedirs(output_dir, exist_ok=True)

carbon_df.to_json(os.path.join(output_dir, 'carbon_example.json'))
```
